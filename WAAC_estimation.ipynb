```python
#Importing the required libraries.
!pip install yfinance
import matplotlib.pyplot as plot
import statsmodels.api as stm
import yfinance as yfin
import pandas as pds
import numpy as npy
import seaborn as sbn
import scipy.stats as stat
```

```python
#Using the stock prices of monthly data of security for last 5 years
sbin_df = yfin.download('SBIN.NS',
                      start='2017-11-01',
                      end='2022-11-01',
                      progress=False,interval ='1mo'
)
sbin_df.head()
```

```python
#Downloading the closing values for the Nifty index
niftybse_df = yfin.download('^BSESN',start='2017-11-01',
                      end='2022-11-01',
                      progress=False,interval ='1mo',
)
niftybse_df.head()
```

```python
# Determining the log returns of nifty
niftybse_df['Return_NIFTY'] = npy.log(niftybse_df['Adj Close']/niftybse_df['Adj Close'].shift(1))
niftybse_df.head()
niftybse_df1=niftybse_df.dropna(axis=0)
niftybse_df1.head()
```

```python
# Calculating the log returns for sbi bank
sbin_df['Return_Sbi']=npy.log(sbin_df['Adj Close']/sbin_df['Adj Close'].shift(1))
sbin_df.head()
sbin_df1 = sbin_df.dropna(axis=0)
sbin_df1.head()
```

```python
# Placing the returns of SBI and NIFTY into columns
sbin_df1_returns = sbin_df1['Return_Sbi']
NIFTY_df1_returns = niftybse_df1['Return_NIFTY']
# Merging the columns of returns
returns= pds.merge(sbin_df1_returns,NIFTY_df1_returns,how='inner',on='Date')
returns.head()
returns.describe()
```

- *Annualised Returns_SBI = 12.35% ( We are taking APR across all returns, so Annualised Returns are calculated as 12 * mean Returns from above table)*
- *Annualised Returns_NIFTY = 12.32%*

Market return(Rm)=12.32%


```python
# Declaring Regression Variables
x=returns['Return_NIFTY']
y=returns['Return_Sbi']
x_sm=stm.add_constant(x)
```

```python
#Regression
model = stm.OLS(y,x_sm)
#OLS estimation or Ordinary Least square Minimizes the sum of squared errors
results = model.fit()
results.summary()
```

# **Calculating the Cost of Debt:**

```python
#To calculate the interest rates(pre-tax cost of debts) according to the estimated ratings
def rating_spread(int_cov):
    #For the firms with market capitalisation > $5billion
    if int_cov > 8.5:
        #Rating is AAA
        creditrating_spread = 0.0069
    elif (int_cov > 6.5) & (int_cov <= 8.5):
        #Rating is AA
        creditrating_spread = 0.0085
    elif (int_cov > 5.5) & (int_cov <= 6.5):
        #Rating is A+
        creditrating_spread = 0.0107
    elif (int_cov > 4.25) & (int_cov <= 5.49):
        #Rating is A
        creditrating_spread = 0.0118
    elif (int_cov > 3) & (int_cov <= 4.25):
        #Rating is A-
        creditrating_spread = 0.0133
    elif (int_cov > 2.5) & (int_cov <= 3):
        #Rating is BBB
        creditrating_spread = 0.0171
    elif (int_cov > 2.25) & (int_cov <= 2.5):
        #Rating is BB+
        creditrating_spread = 0.0231
    elif (int_cov > 2) & (int_cov <= 2.25):
        #Rating is BB
        creditrating_spread = 0.0277
    elif (int_cov > 1.75) & (int_cov <= 2):
        #Rating is B+
        creditrating_spread = 0.0405
    elif (int_cov > 1.5) & (int_cov <= 1.75):
        #Rating is B
        creditrating_spread = 0.0486
    elif (int_cov > 1.25) & (int_cov <= 1.5):
        #Rating is B-
        creditrating_spread = 0.0594
    elif (int_cov > 0.8) & (int_cov <= 1.25):
        #Rating is CCC
        creditrating_spread = 0.0946
    elif (int_cov > 0.65) & (int_cov <= 0.8):
        #Rating is CC
        creditrating_spread = 0.0997
    elif (int_cov > 0.2) & (int_cov <= 0.65):
        #Rating is C
        creditrating_spread = 0.1309
    elif int_cov <= 0.2:
        #Rating is D
        creditrating_spread = 0.1744
    return creditrating_spread
```
